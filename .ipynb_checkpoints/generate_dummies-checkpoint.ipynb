```python
import json
import pandas as pd
import os

# --- KONFIGURASI ---
GEOJSON_PATH = "data/kecamatan.geojson"
OUTPUT_DIR = "data"
START_DATE = "2024-01-01"
END_DATE = "2024-03-31"
FIXED_VOLUME = 4000  # <--- NILAI TETAP

def get_kecamatan_from_geojson():
    """Membaca daftar kecamatan dari GeoJSON untuk menjamin kecocokan 100%."""
    if not os.path.exists(GEOJSON_PATH):
        print(f"âŒ Error: File {GEOJSON_PATH} tidak ditemukan.")
        return []
    
    print(f"ğŸ“‚ Membaca GeoJSON dari {GEOJSON_PATH}...")
    with open(GEOJSON_PATH, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    kecamatan_list = set()
    for feature in data.get('features', []):
        props = feature.get('properties', {})
        name = props.get('district') or props.get('kecamatan') or props.get('name')
        if name:
            # Normalisasi standar (UPPERCASE + Hapus Non-breaking space)
            clean_name = name.strip().upper().replace('\xa0', ' ')
            kecamatan_list.add(clean_name)
            
    return sorted(list(kecamatan_list))

def generate_sipsn(kecamatan_list):
    """Membuat data profil kecamatan (SIPSN)."""
    print("ğŸ› ï¸  Generating sipsn.csv...")
    data = []
    for kec in kecamatan_list:
        data.append({
            "kecamatan": kec,
            "armada_total": 4000,         # Diset 4000 juga
            "armada_operasional": 4000,   # Diset 4000 juga
            "ritase_harian": 4000,
            "kapasitas_m3": 4000,
            "penduduk": 400000,           # Populasi dummy
            "luas_km2": 40.00
        })
    pd.DataFrame(data).to_csv(f"{OUTPUT_DIR}/sipsn.csv", index=False)
    print("âœ… sipsn.csv generated (All 4000).")

def generate_waste_flat(kecamatan_list):
    """
    Membuat data transaksi sampah harian dengan nilai FLAT 4000.
    """
    print(f"ğŸ› ï¸  Generating waste.csv (Flat {FIXED_VOLUME})...")
    
    date_range = pd.date_range(start=START_DATE, end=END_DATE)
    data = []
    
    for date in date_range:
        for kec in kecamatan_list:
            data.append({
                "tanggal": date.strftime("%Y-%m-%d"),
                "kecamatan": kec,
                "volume_ton": FIXED_VOLUME, # <--- 4000 SEMUA
                "jenis_sampah": "Domestik",
                "sumber_sampah": "Rumah Tangga"
            })
            
    pd.DataFrame(data).to_csv(f"{OUTPUT_DIR}/waste.csv", index=False)
    print(f"âœ… waste.csv generated (Setiap baris = {FIXED_VOLUME}).")

if __name__ == "__main__":
    if not os.path.exists(OUTPUT_DIR):
        os.makedirs(OUTPUT_DIR)
        
    kec_list = get_kecamatan_from_geojson()
    
    if kec_list:
        generate_sipsn(kec_list)
        generate_waste_flat(kec_list)
        print(f"\nğŸ‰ Data Dummy FLAT {FIXED_VOLUME} Berhasil Dibuat!")
    else:
        print("âŒ Gagal. GeoJSON tidak ditemukan.")
```
